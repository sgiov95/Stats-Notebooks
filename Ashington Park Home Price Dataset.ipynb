Real Estate Data typically features a bundle of physical properties or price variables, so they are typically positive and correlate positively. This can yield some interesting linear relationships between the variables, but in the case of one or more independent covariates, might result in serious Collinearity Issues. If we want to predict more than one variable, we also have to take this high degree of correlation into account also. 

There are several different ways to impute the dataset, since it doesn't appear like the data is missing at random. The first method involves filling in missing data fields with the column mean, and is overall the easiest to implement:

```python
df1 = homes.copy()
df1 = df1.fillna(df1.mean())
df1.info()
```

```python
df1.describe()
```

The error / warning appears from the non-numerical column, 'Notes', which has most rows missing. The main thing that would throw off that kind of imputation would be if there are bimodal distributions with missing data, or if there is a categorical factor strongly affecting the price of the homes, such as the number of stories:

```python
homes.groupby('Stories')['ZestimateOct2022'].describe()
```

```python
df1.groupby('Stories')['ZestimateOct2022'].describe()
```

To visualize the missing data and how it is missing across rows and columns, we can make a missingness map:

```python
sns.heatmap(homes.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

In this case, it definitely does not appear that the missing data is missing at random- this is hazardous to any inference that might be made or any models that might be built from the dataset, because since the data being missing is not independent of the data itself, an argument can be made for any conclusions reached from merely dropping any rows that have missing columns that aren't 'Notes' will induce bias in both inference and models.

Instead, we can try to use relations present in the data itself to make some inferences and rely on a pre-hoc analysis to perform an imputation based on the available data. This can still be wildly incorrect, but the data tends to be less wrong if the factors that matter for our pricing model and other context clues are used on the data. For example, there are two columns here without much variance- 'OCPALandValue2021' and 'OCPALandValue2022'. These two columns are only missing in one row, which is missing a bunch of the columns starting with 'OCPA'. However, because the value appears to be the same as the other values, we can impute this as the average value:

```python
df2 = homes.copy()
scols1 = ['OCPALandValue2021' , 'OCPALandValue2022']
df2[scols1] = df2[scols1].fillna(df2.mean())
sns.heatmap(df2.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

Ok, so we need to figure out the values where these things are missing first, and then apply what we need to. So to do this, we can create a temporary object for the imputation:

```python
dftemp1 = df2[scols3]
dftemp1
```

```python
X_pred = dftemp1[dftemp1.isna().sum(axis=1) > 0][scols3[:3]].values.reshape(-1,3)
Y_obs = dftemp1[dftemp1.isna().sum(axis=1) > 0][scols3[3:]].values.reshape(-1,3)
indx = np.where(dftemp1.isna().sum(axis=1) > 0)[0]
```

```python
dftemp1 = dftemp1.dropna()
dftemp1
```

Now, we can easily do this:

```python
X = dftemp1[scols3[:3]].values.reshape(-1,3)
u = np.ones(X.shape[0],dtype=int).reshape(-1,1)
X = np.concatenate((u,X),axis=1)
X
```

```python
Y = dftemp1[scols3[3:]].values.reshape(-1,3)
Y
```

```python
print(X.shape,Y.shape)
```

Perfect. So first thing that we can do here is find the regression coefficients for the best fit MANOVA:

```python
A = np.matmul(X.T,X)
C = np.matmul(X.T,Y)
B = np.matmul(np.linalg.inv(A),C)
B
```

So there is the straightforward way to interpret these numbers here, and it is that, for example, for every 1 sq ft increase in the living area of a house, there is on average, a 49.48 USD increase in the ZestimateOct2022 of the house, with similar increases for the RealtorcomEstimate and TruliaEstimate columns. So we can use this to get a bunch of predictions, but what we can also do is look back at the dataset and use that to fill in the columns that are missing there:

```python
u = np.ones(X_pred.shape[0],dtype=int).reshape(-1,1)
X_pred = np.concatenate((u,X_pred),axis=1)
X_pred
```

```python
Y_pred = np.matmul(X_pred,B)
Y_pred
```

```python
j = 0
for i in indx:
    if np.isnan(df2.at[i,scols3[3]]) == True:
        df2.at[i,scols3[3]] = Y_pred[j,0]
    if np.isnan(df2.at[i,scols3[4]]) == True:
        df2.at[i,scols3[4]] = Y_pred[j,1]
    if np.isnan(df2.at[i,scols3[5]]) == True:
        df2.at[i,scols3[5]] = Y_pred[j,2]
    j += 1

df2[scols3].loc[indx]
```

```python
X = df2.Sqft.values.reshape(-1,1)
u = np.ones(len(df2),dtype=int).reshape(-1,1)
X = np.concatenate((u,X),axis=1)
Y = df2.ZestimateOct2022.values.reshape(-1,1)
A = np.linalg.inv(np.matmul(X.T,X))
C = np.matmul(X.T,Y)
B = np.matmul(A,C)
B
```

```python
Y_pred = np.matmul(X,B)
Y_pred-Y
```

```python
SST = np.matmul((Y-Y.mean(axis=0)).T,Y-Y.mean(axis=0))
SST
```

```python
SSE = np.matmul((Y-Y_pred).T,Y-Y_pred)
SSE
```

```python
rsq = 1-(SSE/SST)
rsq
```

```python
MSE = np.sqrt(SSE/(len(df2)-2))
MSE
```

Ok, so let's do a T test, 99%, for the coefficient estimate $\beta_1$:

```python
mux = X.mean(axis=0)[1]
sem_b1 = MSE/np.sqrt(np.sum((X[:,1]-mux)**2))
sem_b1
```

```python
test_stat = B[1]/sem_b1
test_stat
```

```python
import scipy.stats as ss
alpha=.01
degf = len(df2) - 2
crit_val = ss.t(df=degf).ppf(1-alpha/2)
crit_val
```

So we would reject $H_0 : \beta_1 = 0$ and to get the P-value, we'd get:

```python
pval = 2*(1-ss.t(df=degf).cdf(test_stat))
pval
```

For a corresponding 99% confidence interval for $\beta_1$, we use:

```python
ci99_b1 = B[1] + np.array([-1,1])*crit_val*sem_b1
ci99_b1
```

So a 99% Confidence Interval for $\beta_1$ is between 75.25 and 95.08 USD per Sqft. What about a 99% Confidence Interval for the mean value of a house with a Sqft of 2400: 

```python
x_val = 2400
ssx = np.sum((X-mux)**2)
sy_hat_x = MSE*np.sqrt((1/len(df2)) + ((x_val-mux)**2)/ssx)
sy_hat_x
```

```python
ci99_muyx = Y.mean() + B[1]*(x_val - mux) + np.array([-1,1])*crit_val*sy_hat_x
ci99_muyx
```

So a 99% CI for the ZestimateOct2022 for the mean value of a home in this neighborhood that has 2400 sqft is between 446.8k and 454.3k USD. What about a prediction interval ?

```python
s_predyx = MSE*np.sqrt(1+(1/len(df2)) + ((x_val-mux)**2)/ssx)
pi99_yx = Y.mean() + B[1]*(x_val - mux) + np.array([-1,1])*crit_val*s_predyx
pi99_yx
```

So a 99% Prediction Interval for the ZestimateOct2022 for the value of a home in this neighborhood that has 2400 sqft is between 416.9k and 484.2k USD. We can compute PI's for each observation:

```python
lowr = np.zeros(len(df2))
uppr = np.zeros(len(df2))
outside = np.zeros(len(df2))
for i in range(len(df2)):
    x_val = df2.at[i,'Sqft']
    s_predyx = MSE*np.sqrt(1+(1/len(df2)) + ((x_val-mux)**2)/ssx)
    lowr[i] = Y.mean() + B[1]*(x_val - mux) - crit_val*s_predyx
    uppr[i] = Y.mean() + B[1]*(x_val - mux) + crit_val*s_predyx
    y_val = df2.at[i,'ZestimateOct2022']
    if (y_val > uppr[i]):
        outside[i] = 1
    elif (y_val < lowr[i]):
        outside[i] = -1
```
